```python
import pandas as pd
from yahooquery import Ticker
```

```python
my_etf_df = pd.read_csv('fixed_esg_etf.csv')
```

```python
t = Ticker('BND')
```

```python
t.fund_sector_weightings.reset_index()
```

```python
my_weig_list = []
for i in range(0, 3171):
    ticker_symbol = my_etf_df.ticker[i]
    try:
        t = Ticker(ticker_symbol)
        tickerinfo_df = t.fund_sector_weightings
        my_weig_list.append(tickerinfo_df)
    except Exception as e:
        print(f"Error for ticker {ticker_symbol}: {e}")

print(my_weig_list)
```

```python
len(my_weig_list)
```

```python
my_weig_list_concat = pd.concat(my_weig_list,axis=1)
```

```python
my_weig_list_concat
```

```python
my_weig_list_concat.to_csv('my_weig_list_concat.csv')
```
