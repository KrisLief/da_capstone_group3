### drop ticker for imputing

```python
numericals_trunc = numericals.drop('ticker', axis=1)
numericals_trunc.columns
```

### impute nan by KNN

```python
from sklearn.impute import KNNImputer

# Assuming your DataFrame is named 'df'
# If your DataFrame is not already loaded, load it first.

# Create the KNNImputer with the desired number of neighbors (e.g., 5)
imputer = KNNImputer(n_neighbors=5)

# Fit and transform the DataFrame to impute the NaN values
numericals_trunc_imputed = pd.DataFrame(imputer.fit_transform(numericals_trunc), columns=(numericals_trunc.columns))
```

```python
numericals_trunc_imputed.isna().sum()
```

```python
numericals_trunc.describe()
```

```python
numericals_trunc_imputed.info()
```

### reattach ticker on imputed data frame

```python

def replace_columns(df1, df2):
    # Create a new DataFrame using the first column of df1
    df3 = pd.DataFrame(df1[df1.columns[0]])

    # Replace all other columns of df1 with the corresponding columns from df2
    df3[df2.columns] = df2

    return df3

```

```python
numericals_imputed = replace_columns(numericals, numericals_trunc_imputed)
numericals_imputed.head()
```

```python
numericals_imputed.isna().sum()
```

### merge numericals and categoricals on ticker

```python
df_input = pd.merge(numericals_imputed, categoricals, how='left', on= 'ticker')
df_input.head()
```

### scale all numeric columns of df_input and transform to array

```python
from sklearn.preprocessing import StandardScaler

# Separate numeric and non-numeric columns
numeric_cols = df_input.select_dtypes(include=[np.number]).columns

# Create a StandardScaler object
scaler = StandardScaler()

# Fit and transform the numeric columns to scale/normalize the features
df_input[numeric_cols] = scaler.fit_transform(df_input[numeric_cols])
```

### create umap embeddings from input

```python
import umap

# Create a UMAP object with desired hyperparameters
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=3)

# Fit the UMAP model to the scaled data to obtain embeddings
umap_embeddings = umap_model.fit_transform(df_input[numeric_cols])

# Create a DataFrame to store the UMAP embeddings
umap_data = pd.DataFrame(umap_embeddings, columns=['UMAP1', 'UMAP2', 'UMAP3'])
```
